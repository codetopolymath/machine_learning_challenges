```python
import matplotlib.pyplot as plt 
import pandas as pd
data = pd.read_csv("./input/Demand_History.csv")

x = data.iloc[:, 1].values
y = data.iloc[:, 2].values

plt.scatter(x,y, s=0.1, c='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('DEMAND MATRIX OVER TIME')
plt.show()
```

```python
demand = pd.read_csv("./input/Demand_History.csv")
supply = pd.read_csv("./input/exisiting_EV_infrastructure_2018.csv")

x1 = demand.iloc[:, 1].values
y1 = demand.iloc[:, 2].values
x2 = supply.iloc[:, 1].values
y2 = supply.iloc[:, 2].values


fig, ax = plt.subplots()

# Create the first scatter plot
ax.scatter(x1, y1, s=0.1, c='blue')

# Create the second scatter plot
ax.scatter(x2, y2, c='red')

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('DEMAND SUPPLY MATRIX')

# Show the plot
plt.show()
```
